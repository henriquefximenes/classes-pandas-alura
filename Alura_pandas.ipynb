1. Quais os valores médios de aluguel por tipo de imóvel?
2. Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
dados['Valor'].mean() #aprendendo a utilizar o método mean()
```

Agora, esse valor não é exatamente o que queremos. Nós queremos o valor por tipo de imóvel. Para isso temos o método groupby()

```python
dados.groupby('Tipo').mean(numeric_only=True)
```

Podemos ver que ainda não é o que queremos, pois agrupou por Tipo, mas apresentou a média de todos os valores numéricos. Queremos o valor do imóvel somente.

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor') #apresenta os dados de forma ordenada com o crescimento do valor.
```

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_preco_tipo.plot(kind='barh', figsize=(7, 7), color='red', title='Valores Médios de Aluguel') #.plot é o método para plotar gráficos com a biblioteca Pandas.
```

Fazendo o .sum() em conjunto com .isnull() podemos identificar quais colunas possuem dados nulos e, inclusive, obtemos a soma da quantidade de dados nulos por coluna.

```python
df.dropna() #este método retira todos os dados nulos do dF.
```

Foi possível realizar esse tipo de método por termos somente valores nulos em colunas de dados numéricos. Em caso de outro tipo de situação, teríamos de identificar melhor como tratar os dados.

```python
df = df.fillna(0) #este método adiciona valor para as células que possuem valores nulos.
```

Agora, que retiramos os valores nulos e inserimos valores '0' (zero) no lugar, podemos remover registros inconsistentes da base de dados do dataFrame tratado. Para isso iremos retirar valores zerados do dF.

```python
df.query('Valor == 0 | Condominio == 0') #trará como resultado todas as linhas onde Valor=0 ou Condominio=0
```

```python
df.query('Valor == 0 | Condominio == 0').index
```

```python
registros_remover = df.query('Valor == 0 | Condominio == 0').index
```

```python
df.drop(registros_remover, axis=0, inplace=True)
'''Tem-se 3 considerações neste código:
      1. drop() utilizamos para retirar linhas ou colunas desejadas
      2. axis=0 significa que queremos retirar a linha - se tivesse =1 seria a coluna
      3. inplace=True utilizamos para não precisar realizar a igualdade com o próprio dF.'''
```

```python
df.query('Valor == 0 or Condominio == 0') # conferindo se conseguimos remover os registros zerados.
```

```python
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
#Criando uma coluna nova valor/mês
dados['valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['valor_por_ano'] = dados['valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

Podemos observar que conseguimos criar colunas novas, realizar operações entre elas e, também, executar operções por escalar.

```python
dados['descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] #concatenando as informações string
dados.head()
```

```python
dados['descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' +\
dados['Quartos'] + ' quarto(s) ' +\
 ' e ' + dados['Vagas'] + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' +\
                      dados['Quartos'].astype(str) + ' quarto(s) ' +\
                      ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
#utilizando função lambda para criar uma função e criar a coluna com sim e não.
dados['possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index = False, sep=';')
```
